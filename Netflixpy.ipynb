```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('C:\\Users\\PASSNOWNOW\\Documents\\netflix.csv')
```

```python
df.head()
```

```python
# checking for missing values 
df.isnull().sum()
```

```python
# drop unwanted columns
df.drop(['director','cast','country'],axis =1,inplace =True)
```

```python
df.head()
```

```python
# check missing values
df.isnull().sum()
```

```python
# drop Missing Values
df.dropna(inplace=True)
```

```python
#  Analyze the distribution of ratings
rating_counts = df['rating'].value_counts()
print(rating_counts)
```

```python
# Extract genres and count occurrences
genre_counts = df['listed_in'].str.split(', ').explode().value_counts()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.xticks(rotation=90)
plt.title('Most Watched Genres on Netflix')
plt.xlabel('Genres')
plt.ylabel('Count')
plt.show()
```

```python
#Ratings Distribution
plt.figure(figsize=(10, 5))
sns.countplot(x='rating', data=df)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
